# Data Visualization
Data visualization is the practice of translating information into a visual context, such as a map or graph, to make data easier for the human brain to understand and pull insights from. The main goal of data visualization is to make it easier to identify patterns, trends and outliers in large data sets, which is essential for exploratory data analysis (EDA). In this notebook, some codes for producing visualizations using popular Python libraries such as matplotlib, seaborn, pandas built-in and plotly are provided.

# Import relevant libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('whitegrid')
```

# Data visualization using matplotlib

Import seaborn's built-in dataset 'titanic'

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

Histogram

A histogram is a graph that shows the frequency of numerical data using rectangles. The height of a rectangle (the vertical axis) represents the distribution frequency of a variable (the amount, or how often that variable appears).

```python
# Plot histogram for age
plt.hist(x='age',data = titanic,bins=20)

plt.title("Histogram for Age", fontsize = 20)
plt.xlabel("Age",fontsize = 15)
plt.ylabel("Count",fontsize = 15)

plt.show()
```

```python
# Plot histogram for fare
plt.hist(x='fare',data = titanic,bins=20)

plt.title("Histogram for Fare", fontsize = 20)
plt.xlabel("Fare",fontsize = 15)
plt.ylabel("Count",fontsize = 15)

plt.show()
```

Bar Chart

A bar chart is a way of summarizing a set of categorical data. The bar chart displays data using a number of bars, each representing a particular category.

```python
titanic['sex'].unique()
```

```python
titanic['sex'].value_counts()
```

```python
# Plot a bar chart for sex
x_labels = list(titanic['sex'].unique())
y_labels = list(titanic['sex'].value_counts())

plt.bar(x = x_labels,height=y_labels)

plt.title("Bar Chart for Sex", fontsize = 20)
plt.xlabel("Sex", fontsize=15)
plt.ylabel("Count", fontsize=15)

plt.show()
```

```python
# Plot a bar chart for passenger class
x_labels = list(titanic['pclass'].unique())
y_labels = list(titanic['pclass'].value_counts())

plt.bar(x = x_labels,height=y_labels)

plt.title("Bar Chart for Passenger Class", fontsize = 20)
plt.xlabel("Passenger Class", fontsize = 15)
plt.ylabel("Count",fontsize = 15)

plt.show()
```

Scatter Plot

A scatter plot (also called a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram) is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. Scatter plot is commonly used for determining if there is any relationship between the two variables and visualizing potential outliers.

```python
# Plot a scatter plot for fare vs age

plt.scatter(x = 'age', y = 'fare', data = titanic)

plt.title("Scatter Plot for Fare vs Age", fontsize = 20)
plt.xlabel("Age", fontsize=15)
plt.ylabel("Fare", fontsize=15)

plt.show()
```

```python
titanic['survived'].unique()
```

```python
titanic['survived'].value_counts()
```

Pie Chart

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area) is proportional to the quantity it represents. 

```python
# Plot a pie chart for survival (in percentages)
labels = titanic['survived'].unique()
sizes = titanic['survived'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(sizes,autopct='%.0f%%',textprops={'fontsize': 15})
plt.title("Pie Chart for Survival", fontsize = 20)
plt.legend(['Survived','Not Survived'],bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, fontsize = 15)

plt.axis('equal')
plt.show()
```

```python
# Plot a pie chart for survival (in absolute values)
labels = titanic['survived'].unique()
sizes = titanic['survived'].value_counts()

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)
    return a

plt.figure(figsize=(5,5))
plt.pie(sizes,autopct=absolute_value,textprops={'fontsize': 15})
plt.title("Pie Chart for Survival", fontsize = 20)
plt.legend(['Survived','Not Survived'],bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0,fontsize = 15)

plt.axis('equal')
plt.show()
```

# Data visualization using seaborn

Import seaborn's built-in dataset 'tips'

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

Plots for studying distribution of dataset: histplot, jointplot, pairplot

histplot is used for showing the distribution of a univariate set of observations

```python
# Plot a hisplot for total_bill
sns.histplot(tips['total_bill'],kde=True)

plt.title("Histplot for Total Bill", fontsize = 20)
plt.xlabel("Total Bill", fontsize=15)
plt.ylabel("Count", fontsize=15)

plt.show()
```

```python
# Adjust the bins of the histplot to 25
sns.histplot(tips['total_bill'],kde=True,bins=25)

plt.title("Histplot for Total Bill", fontsize = 20)
plt.xlabel("Total Bill", fontsize=15)
plt.ylabel("Count", fontsize=15)

plt.show()
```

jointplot allows us to basically match up two distplots for bivariate data.

```python
# Plot a jointplot for tip vs total_bill (kind = 'scatter')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')

plt.show()
```

```python
# Plot a jointplot for tip vs total_bill (kind = 'hex')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')

plt.show()
```

```python
# Plot a jointplot for tip vs total_bill (kind = 'reg')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')

plt.show()
```

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a

```python
# Plot a pairplot for tips dataset
sns.pairplot(tips)

plt.show()
```

```python
# Plot a pairplot for tips dataset, with colour differentiated by sex
sns.pairplot(data = tips,hue='sex',palette='coolwarm')

plt.show()
```

Plots for categorical data

```python
tips.head()
```

barplot is a general plot that allows us to aggregate the categorical data based off some function, by default the mean

```python
# Plot a barplot for sex, with bar height equal to mean of total_bill
sns.barplot(x='sex',y='total_bill',data=tips)

plt.show()
```

```python
# Plot a barplot for sex, with bar height equal to standard deviation of total_bill
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)

plt.show()
```

countplot is used for showing the number of occurrence for each category

```python
# Plot a countplot for sex
sns.countplot(x='sex',data=tips)

plt.show()
```

Boxplot and violinplot

Boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.

Violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which
all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the
underlying distribution.

```python
# Plot a boxplot for total_bill vs day
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')

plt.show()
```

```python
# Plot a boxplot for total_bill vs day, with colour differentiated by whether the customer is smoker
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.show()
```

```python
# Plot a violinplot for total_bill vs day
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')

plt.show()
```

```python
# Plot a violinplot for total_bill vs day, with colour differentiated by sex
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.show()
```

```python
# Plot a violinplot for total_bill vs day, with colour differentiated by sex
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.show()
```

Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

Import another built-in seaborn dataset 'iris'

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# Plot the PairGrid for iris dataset
g = sns.PairGrid(iris)
g.map(plt.scatter)

plt.show()
```

```python
# Plot the PairGrid for iris dataset, with colour differentiated by species
g = sns.PairGrid(iris,hue = 'species')
g.map(plt.scatter)

plt.legend(bbox_to_anchor=(1.2, 4.25), loc='upper left', borderaxespad=0)
plt.show()
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

plt.show()
```

pairplot

pairplot is a simpler version of PairGrid (we will use it quite often)

```python
# Plot the pairplot for iris dataset
sns.pairplot(iris)

plt.show()
```

```python
# Plot the pairplot for iris dataset, with colour differentiated by species
sns.pairplot(iris,hue='species',palette='rainbow')

plt.show()
```

Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature

Using back tips dataset

```python
tips.head()
```

```python
# Produce histograms for total_bill in FacetGrid as conditioned by smoker (row) and time (column)
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")

plt.show()
```

```python
# Produce scatter plots for tip vs total_bill in FacetGrid as conditioned by smoker (row) and time (column),
# with colour differentiated by sex
g = sns.FacetGrid(tips, col="time", row="smoker",hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()

plt.show()
```

JointGrid

JointGrid is the general version for jointplot() type grids

```python
# Produce a JointGrid plot for tip vs total_bill using sns.regplot and sns.histplot
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.histplot)

plt.show()
```
